# Psychological Trigger Definitions

Curiosity:
Creates an information gap.

Opportunity:
Creates desire for gain.

Tutorial:
Promises practical learning.

Fear:
Creates concern about negative outcomes.

Authority:
Signals expertise or success.

Loss Avoidance:
Promises protection from mistakes.

Future Uncertainty:
Focuses on future events or predictions.

Urgency:
Suggests immediate action is required.

```python
def contains_any_word(title, words):
    title = str(title).upper()
    return any(word in title for word in words)

df["fear_title"] = df["title"].apply(
    lambda x: contains_any_word(x, fear_words)
)

df["curiosity_title"] = df["title"].apply(
    lambda x: contains_any_word(x, curiosity_words)
)

df["opportunity_title"] = df["title"].apply(
    lambda x: contains_any_word(x, opportunity_words)
)

df["fear_score"] = df["fear_title"].astype(int)
df["curiosity_score"] = df["curiosity_title"].astype(int)
df["opportunity_score"] = df["opportunity_title"].astype(int)
```

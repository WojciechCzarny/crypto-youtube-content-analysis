```python
import pandas as pd

df = pd.read_csv(
    "data/processed/master_dataset_with_title_features.csv"
)

print(df.shape)
```

```python
df.columns.tolist()
```

```python
def contains_any_word(title, words):
    title = str(title).upper()
    return any(word in title for word in words)
```

```python
exchange_words = [
    "BINANCE",
    "BYBIT",
    "ZONDA"
]

tutorial_words = [
    "JAK",
    "PORADNIK",
    "KURS",
    "USTAWIĆ",
    "UŻYWAĆ"
]
```

```python
df["exchange_topic"] = df["title"].apply(
    lambda x: contains_any_word(x, exchange_words)
)

df["tutorial_title"] = df["title"].apply(
    lambda x: contains_any_word(x, tutorial_words)
)
```

```python
df[
    [
        "exchange_topic",
        "tutorial_title"
    ]
    ].sum()
```

```python
df["curiosity_exchange"] = (
    (df["curiosity_title"] == True)
    &
    (df["exchange_topic"] == True)
)
```

```python
df["curiosity_exchange"].value_counts()
```

```python
df.groupby("curiosity_exchange")[
    [
        "view_count",
        "comment_count",
        "engagement_rate"
    ]
    ].mean()
```

```python
df.to_csv(
    "data/processed/master_dataset_with_content_formulas.csv",
    index=False,
    encoding="utf-8-sig"
)
```

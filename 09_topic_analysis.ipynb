```python
import pandas as pd

df = pd.read_csv(
    "data/processed/master_dataset_with_title_features.csv"
)

print(df.shape)
```

```python
df.columns.tolist()
```

```python
def contains_any_word(title, words):
    title = str(title).upper()
    return any(word in title for word in words)
```

```python
df["bitcoin_topic"] = df["title"].apply(
    lambda x: contains_any_word(x, bitcoin_words)
)

df["altcoin_topic"] = df["title"].apply(
    lambda x: contains_any_word(x, altcoin_words)
)

df["trading_topic"] = df["title"].apply(
    lambda x: contains_any_word(x, trading_words)
)

df["exchange_topic"] = df["title"].apply(
    lambda x: contains_any_word(x, exchange_words)
)

df["macro_topic"] = df["title"].apply(
    lambda x: contains_any_word(x, macro_words)
)
```

```python
df[
    [
        "bitcoin_topic",
        "altcoin_topic",
        "trading_topic",
        "exchange_topic",
        "macro_topic"
    ]
].sum()
```

```python
df.groupby("bitcoin_topic")[
    [
        "view_count",
        "comment_count",
        "engagement_rate"
    ]
].mean()
```

```python
topic_columns = [
    "bitcoin_topic",
    "altcoin_topic",
    "trading_topic",
    "exchange_topic",
    "macro_topic"
]

for topic in topic_columns:
    print("\n" + topic)
    display(
        df.groupby(topic)[
            [
                "view_count",
                "comment_count",
                "engagement_rate"
            ]
        ].mean()
    )
```
